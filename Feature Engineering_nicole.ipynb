```python
numerical_features = ['GrLivArea', 'LotFrontage', 'LotArea', 'YearBuilt', 'YearRemodAdd','MasVnrArea', 
                     'BsmtFinSF1', 'BsmtFinSF2', 'BsmtUnfSF', 'TotalBsmtSF', '1stFlrSF','2ndFlrSF', 
                     'GarageArea','WoodDeckSF','TotalPorchSF']

ordinal_features = ['OverallQual', 'OverallCond','MSSubClass','ExterQual','ExterCond','BsmtQual','BsmtCond','BsmtExposure',
                'BsmtFinType1','HeatingQC','BedroomAbvGr','KitchenQual','TotRmsAbvGrd','Fireplaces',
                'FireplaceQu','GarageCars', 'GarageQual', 'GarageCond', 'MoSold', 'YrSold', 'MSSubClass','TotalBR']

cat_features = ['MSZoning', 'LotShape', 'Neighborhood','BldgType','HouseStyle', 'PavedDrive','Exterior1st','MasVnrType','Foundation','CentralAir', 'Electrical','GarageType', 'GarageFinish']
```

```python
df_housing[numerical_features] = df_housing[numerical_features].astype('float')
```

#### Combining related & dependent features

```python
#combining number of bathrooms 
#turning half baths to .5
df_housing['BsmtHalfBath'] = df_housing['BsmtHalfBath'].replace([1], 0.5)
df_housing['HalfBath'] = df_housing['HalfBath'].replace([1], 0.5)

#adding half bathrooms to full baths
df_housing['BsmtBR'] = df_housing['BsmtHalfBath'] + df_housing['BsmtFullBath']
df_housing['AbvBR'] = df_housing['HalfBath'] + df_housing['FullBath']
df_housing['TotalBR'] = df_housing['AbvBR'] + df_housing['BsmtBR']

df_housing = df_housing.drop(['BsmtHalfBath','BsmtFullBath','HalfBath','FullBath','AbvBR','BsmtBR'], axis = 1)

df_housing.head(2)
```

#### Ordinal Features & Label Encoding
 - converting ordinal categories using integer encoding


```python
# converting ordinal categories using integer encoding

ordinal_features_encode = ['OverallQual', 'OverallCond','ExterQual','ExterCond','BsmtQual','BsmtCond','BsmtExposure',
                'BsmtFinType1','HeatingQC','BedroomAbvGr','KitchenQual','TotRmsAbvGrd','Fireplaces',
                'FireplaceQu','GarageCars', 'GarageQual', 'GarageCond']

dict_1 = {'Ex':5, 'Gd':4, 'TA':3, 'Fa':2,'Po':1, 'None':0} #9
dict_2 = {'Gd':4, 'Av':3, 'Mn':2, 'No':1,'None':0}
dict_3 = {'GLQ':6,'ALQ':5,'BLQ':4,'Rec':3,'LwQ':2,'Unf':1,'None':0}

for col in ordinal_features_encode: 
    unique_vals = df_housing[col].unique()
    if (set(unique_vals).issubset(set(dict_1.keys()))) == True:
        df_housing [col] = df_housing[col].replace(dict_1)
    elif (set(unique_vals).issubset(set(dict_2.keys()))) == True:   
        df_housing [col] = df_housing[col].replace(dict_2)
    elif (set(unique_vals).issubset(set(dict_3.keys()))) == True:   
        df_housing [col] = df_housing[col].replace(dict_3)
        
```

```python
#check
#df_ordinals = df_housing.filter(ordinal_cols_list, axis=1)
#df_ordinals.isnull().sum(axis=0)
```

#### Categorical Features and Dummifying:  <span style="color: red;">*Dummified Dataset Here*</span>

```python
#dummifying categorical features 
dummy = pd.get_dummies(df_housing, prefix= cat_features, 
                                  columns= cat_features, drop_first = True)
dummy = dummy.dropna()
```

```python
dummy.info()
```

```python
dummy.describe()
```

```python
X = dummy.drop(['SalePrice','lat','long'], axis = 1)
y = dummy['SalePrice']
```

```python
#A measure that takes into account both year built and remodeled
df_housing['YrBltAndRemod']= df_housing['YearBuilt'] + df_housing['YearRemodAdd']

#A measure of total SF
df_housing['TotalSF']= df_housing['TotalBsmtSF'] + df_housing['1stFlrSF'] + df_housing['2ndFlrSF']

#A different measure of total SF
df_housing['TotalSF2'] = df_housing['BsmtFinSF1'] + df_housing['BsmtFinSF2'] + df_housing['1stFlrSF'] + df_housing['2ndFlrSF']

#Variable with value of all bathrooms
df_housing['Total_Bathrooms'] = (df_housing['FullBath'] + (0.5 * df_housing['HalfBath']) +
                           df_housing['BsmtFullBath'] + (0.5 * df_housing['BsmtHalfBath']))

#Variable with total porch square foot area
df_housing['Total_porchSF'] = (df_housing['OpenPorchSF'] + df_housing['3SsnPorch'] + df_housing['EnclosedPorch'] +
                         df_housing['ScreenPorch'] +df_housing['WoodDeckSF'])

#Binary variable for presence of certain house features
df_housing['haspool'] = df_housing['PoolArea'].apply(lambda x: 1 if x > 0 else 0)
df_housing['has2ndfloor'] = df_housing['2ndFlrSF'].apply(lambda x: 1 if x > 0 else 0)
df_housing['hasgarage'] = df_housing['GarageArea'].apply(lambda x: 1 if x > 0 else 0)
df_housing['hasbsmt'] = df_housing['TotalBsmtSF'].apply(lambda x: 1 if x > 0 else 0)
df_housing['hasfireplace'] = df_housing['Fireplaces'].apply(lambda x: 1 if x > 0 else 0)

#Product of related quality scores
df_housing['OverallRating'] = df_housing['OverallQual']*df_housing['OverallCond']
```
